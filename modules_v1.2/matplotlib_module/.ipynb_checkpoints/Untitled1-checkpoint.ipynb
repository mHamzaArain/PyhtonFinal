```python
## Charge Vs Time
### Formula
q_vs_t = lambda n: [round(1-(math.exp(-i)), 3) for i in range(1, n+1)]  # [0.632, 0.865, 0.95, 0.982, 0.993]

plt.plot([0, 1, 2, 3, 4], q_vs_t(5), 'go-.' , linewidth=2.0)
sns.set_style("darkgrid")  # darkgrid, whitegrid, dark, white, ticks
plt.title('Charge Vs. Time')
plt.xlabel('time(t)')
plt.ylabel('Charge(q)')

plt.show()
```

```python
## Capacitance voltage vs time 
### Formula
Vc_vs_t = lambda n: [round(1-(math.exp(-i)), 3) for i in range(1, n+1)]  # [0.632, 0.865, 0.95, 0.982, 0.993

plt.plot([0, 1, 2, 3, 4], Vc_vs_t(5), 'go-.' , linewidth=2.0)
sns.set_style("darkgrid")  # darkgrid, whitegrid, dark, white, ticks
plt.title('Capacitance voltage vs time')
plt.xlabel('time(t)')
plt.ylabel('Capacitance Voltage(Vc)')

plt.show()
```

```python
## Current vs constant time capacitance
plt.plot([0, 1, 2, 3, 4], i_vs_Tc(5), 'go-.' , linewidth=2.0)
sns.set_style("darkgrid")  # darkgrid, whitegrid, dark, white, ticks
plt.title('Capacitance voltage vs time')
plt.xlabel('Constant time capacitance(Tc)')
plt.ylabel('Current(I)')

plt.show()
```
